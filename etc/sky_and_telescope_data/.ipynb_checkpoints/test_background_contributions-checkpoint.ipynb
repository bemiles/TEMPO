```python
import numpy as n
from astropy import constants
from astropy.coordinates import SkyCoord
from astroplan import Observer
from astroplan import FixedTarget
from astropy.time import Time
import astropy.units as u
import configparser
import matplotlib.pyplot as plt

#ETC modules
import sky_emission
import sky_transmission
```

```python
# constants
kb = constants.k_B.value #[J/K]
c = constants.c.value    #[m/s]
h = constants.h.value    #[J/s]   
e_charge = constants.e   #[Coloumbs]
microns2m = 1e-6         #[microns/meters]
deg2rad = 0.0174533      #[deg/radians]
sterr = 1 / 206265**2    #rad^2 / arcsec^2
```

```python
def B_lam(T,lams):

    #Planck's Law

    ### inputs ####
    #T - temperature [Kelvin]
    #lams - microns

    ### output ####
    ## B_lam ### [W/ m^2 / m]

    B_lam_a = (2 * h * c**2) / (lams * microns2m) **5
    B_lam_b = n.exp( h * c / (lams * microns2m) / kb / T ) - 1

    B_lam = B_lam_a / B_lam_b

    return B_lam
```

```python
T_sky = 288              #[Kelvin]
T_telescope = 288        #[Kelvin] 
T_ao = 77                #[Kelvin]
T_zodi = 280             #[Kelvin] Ootsubo 1998, 250 Ootsubo 2000
T_optics = 40            #[Kelvin]

#E_sky = vs. wavlength
E_telescope = 0.09       #value for IRIS #https://github.com/oirlab/liger_iris_sim/blob/master/liger_iris_sim/sky/emission.py
E_ao = 0.01              #value for IRIS #https://github.com/oirlab/liger_iris_sim/blob/master/liger_iris_sim/sky/emission.py
#E_zodi =                   # Ootsubo 2000 https://www.sciencedirect.com/science/article/pii/S0273117799011485
```

```python
wavelengths = n.arange(1,14,0.1)

BB_sky = B_lam(T_sky, wavelengths) 
BB_telescope = B_lam(T_telescope, wavelengths) * E_telescope
BB_ao = B_lam(T_ao, wavelengths) * E_ao
BB_zodi = B_lam(T_zod, wavelengths) 
BB_optics = B_lam(T_optics, wavelengths) 


plt.figure('Emission Sources')
plt.plot(wavelengths, BB_sky, label = 'Sky')
plt.plot(wavelengths, BB_zodi, label = 'Zodi Light')
plt.plot(wavelengths, BB_telescope, label = 'Telescope')
plt.plot(wavelengths, BB_ao, label = 'AO system')
plt.plot(wavelengths, BB_optics, 'k-', label = 'optics')
#plt.yscale('log')
plt.ylabel('Flux (W/ m^2 / m)')
plt.xlabel('Wavelength (microns)')
plt.legend()
plt.show()
```
